```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('pokedex.csv', sep=";")
```

```python
df.info()
```

```python
df.head(10)
```

```python
df_groupped = df.groupby('TYPE').agg({
    'NAME':'nunique',
    'TOTAL': 'mean',
    'HP': 'max',
    'ATTACK': 'min'
}).reset_index().rename(columns={'TYPE':'Category', 'NAME': 'No of Pokemons', 'TOTAL': 'Avg Points', 'HP': 'Max HP', 'ATTACK':'Min Attack'})
```

```python
df_groupped.head()
```

```python
df_groupped.to_csv('poke_report_by_category.csv', index=False)
```
